````python
from codenamize import codenamize
from py2neo import Node, Relationship

import random

prefixes = ['RS', 'SDR', 'TEST']
req_map = {}
for prefix in prefixes:
    req_map_by_level = []
    req_map[prefix] = req_map_by_level
    for x in range(100):
        as_map = {}
        nb = random.randint(1,101)
        as_map['code']= 'REQ-'+prefix+'-'+str(x).zfill(3)
        as_map['name'] = codenamize(nb)
        as_map['nature']= prefix
        req_map_by_level.append(as_map)


relations  = []

def shuffle(levelN, levelNMinusOne):
    listN = req_map[levelN]
    listM = req_map[levelNMinusOne]
    for r in listN:
        nb = random.randint(1,4)
        for i in range(0,nb):
            idx = random.randint(0,len(listM)-1)
            relations.append((r['code'],listM[idx]['code']))

shuffle(prefixes[0], prefixes[1])
shuffle(prefixes[1], prefixes[2])

def _create_relation(tup):
    from_node_id, to_node_id = tup
    return '{from_id:"'+from_node_id+'", to_id:"'+to_node_id+'"}'
    
req_relations = [ _create_relation(r) for r in relations]
````

````python
req_nodes = ['{code:"'+r['code']+'", name:"'+r['name']+'", nature:"'+r['nature']+'" }' for prefix in prefixes for r in req_map[prefix]]

':params req => ['+','.join(req_nodes)+']'
````
